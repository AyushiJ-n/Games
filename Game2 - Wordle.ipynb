# Game : Wordle

```python
# Importing Libraries
import random
from colorama import Fore, Back, Style

# Rules of the Game
print("Welcome to wordle !!\nHere are some rules to the game.\n1) You will have 6 attempt to guess the word.")
print("2) It will be a 5 letter word.\n3) Green alphabet means - Correct alphabet and rightly placed.")
print("4) Yellow alphabet means - Correct alphabet but wrongly placed.\n5) Red alphabet means - Incorrect alphabet.")
print("Keep capslock off for better experience.")

# Reading words from text file
file = open("Game2 - wordlist.txt", 'r')
list_of_words = file.read().split(',')
file.close()
word = random.choice(list_of_words)

user_input = ''

# Taking input from user
Guesses = 6
while Guesses:
    Display = ''
    del user_input
    user_input = input(Fore.RESET + "Enter your word : ")
    Guesses -= 1
    print(Fore.RESET + "Remaining Guesses : ", Guesses)

# Finding out which letters are correct and rightly placed
    for i in range(len(user_input)):
        flag = False
        for j in range(len(word)):
            
            # Making that alphabet colored green
            if user_input[i]==word[j] and i==j:
                Display += Fore.GREEN + user_input[i].upper()
                flag = True
            
            #Making that alphabet colored Yellow
            elif user_input[i]==word[j] and i!=j:
                Display += Fore.YELLOW + user_input[i].upper()
                flag = True
        
        # Making that alphabet colored red
        if not flag: 
            Display += Fore.RED + user_input[i].upper() 

# Displaying the user input in right colours
    print('\t\t\t\t\t\t',Display,'\n')
    print(Fore.RESET)
    del Display
    if user_input == word:
        print(Fore.RESET + "You guessed it correct.")
        break

# Displaying he correct answer
print(word.upper(), "is the right answer")
if Guesses == 0 and user_input != word :
    print("Better Luck Next Time !")
```
